### Jak narysować wykres poziomicowy funkcji, a na nim strzałke?

```python
import matplotlib.pyplot as plt
import cec2017
import numpy as np

from cec2017.functions import f1, f2, f3

MAX_X = 100
PLOT_STEP = 0.1

x_arr = np.arange(-MAX_X, MAX_X, PLOT_STEP)
y_arr = np.arange(-MAX_X, MAX_X, PLOT_STEP)
X, Y = np.meshgrid(x_arr, y_arr)
Z = np.empty(X.shape)

q=f1

for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        Z[i, j] = q(np.array([X[i, j], Y[i, j]]))
        
plt.contour(X, Y, Z, 20)
plt.arrow(0, 0, 50, 50, head_width=3, head_length=6, fc='k', ec='k')
plt.show()
```

## Funkcja Booth
![alt text](booth2.png)

```python
def booth(x: np.array) -> np.array:
    return (x[0] + 2 * x[1] - 7) ** 2 + (2 * x[0] + x[1] - 5) ** 2
```

```python
def calc(func: callable, UPPER_BOUND = 10, DIMENSIONALITY = 2, STEP = 0.01):
    # UPPER_BOUND = 10
    # DIMENSIONALITY = 2
    #wylosuj punkt x:
    x = np.random.uniform(-UPPER_BOUND, UPPER_BOUND, size=DIMENSIONALITY)
    
    print(f"Punkt startowy: {x}")
    # STEP =  0.01

    for i in range(1000):
        d = grad(func)(x)
        x = x - STEP * d

        # if i % 10 == 0:
        #     print(f"Iter: {i} x: {x} d: {d} q: {func(x):6f}")

    # print('q(x) = %.6f' %q)
    print(f"Znalezione minimum: {x} wartość funkcji w tym punkcie {func(x)}")
```

```python
calc(booth, 10, 2, 0.01)
```
