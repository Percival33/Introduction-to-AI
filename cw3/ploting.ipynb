````python
def pd_plot(func: int):
    df = data[ data['quality_func'] == func]
    width = 0.3
    df = df.drop(columns=['games_played', 'depth', 'avg_moves', 'quality_func'], axis=1)
    # print(df.columns)
    # return
    
    # color={"speed": "red", "lifespan": "green"}
    ax = df.plot(kind='bar', stacked=True, figsize=(8, 6), rot=0, title=f'Wyniki rozegranych gier dla funkcji {func}', xlabel='uruchomienia', ylabel='rozkład gier', color={"white_won": "olive", "ties": "orange", "blue_won": "blue"})
    # ax.title()
    for c in ax.containers:
        labels = [v.get_height() if v.get_height() > 0 else '' for v in c]
        # print(c)
        # remove the labels parameter if it's not needed for customized labels
        ax.bar_label(c, labels=labels, label_type='center')

    # print(f"{df['games_played'].sum():.2f}, {df['avg_moves'].min():.2f}, {df['avg_moves'].mean():.2f}, {df['avg_moves'].max():.2f}, {df['avg_moves'].std():.2f}")
````

````python
pd_plot(3)
````

````python
def plot_data(func: int):
    df = data[ data['quality_func'] == func]

    ind = np.arange(df.shape[0])
    width = 0.3

    # fig, ax = plt.subplots()

    white_won = plt.bar(ind, df['white_won'], width, color='olive')
    ties = plt.bar(ind, df['ties'], width, bottom=df['white_won'], color='orange')
    blue_won = plt.bar(ind, df['blue_won'], width, bottom=df['ties'], color='blue')

    # avg_moves = plt.plot(df['depth'], df['avg_moves'])
    plt.title(f'Wyniki rozegranych gier z podstawową funkcją oceny {func}')
    plt.ylabel('rozklad gier: biale/remis/niebieskie')
    plt.xlabel('głębokość przeszukiwań')
    plt.xticks(ind, df['depth'])
    plt.legend((white_won[0], ties[0], blue_won[0]), ('białe', 'remisy', 'niebieskie'))
    plt.show()
    print(f"{df['games_played'].sum():.2f}, {df['avg_moves'].min():.2f}, {df['avg_moves'].mean():.2f}, {df['avg_moves'].max():.2f}, {df['avg_moves'].std():.2f}")
````

````python
plot_data(0)
````
